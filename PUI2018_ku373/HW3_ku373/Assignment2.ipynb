```python
url = "https://data.cityofnewyork.us/resource/hrw9-hqf6.json"
```

```python
df = pd.read_json(url)
```

```python
df.head(7)
```

```python
dhs_3c = df[['neversink_date', 'neversink_storage']]
dhs_3c.head(7)
```

```python
dhs_3c.rename(columns = {'neversink_date':'Date', 
               'neversink_storage':'Neversink Storage'}, inplace=True)
```

```python
dhs_3c["Date"] = dhs_3c["Date"].apply(lambda x: pd.Timestamp(x))
ax = dhs_3c.plot(x='Date', y='Neversink Storage')
ax.set_ylabel("Billion Gallons")
plt.title("Amount of Water Storage");
```

Figure 2: This graph shows amount of water storage(Billion Gallons) at Neversink Storage rereported by the Department of Environmental Protection (DEP) between 01/04/2006 and 12/29/2017. 
Amount of the storage has not been stable and had change with some variance.
As it looks the changes happen periodically, it would show the seasonal changes of the water storage.
